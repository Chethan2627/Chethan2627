1) Circle:x^2+y^2=a^2 for a=3

```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(x**2+y**2,9),(x,-9,9),(y,-9,9),line_color='purple',title='Circle:$x^2+y^2=9$')
```

2) Strophoid: y^2(a-x)=x^2(a+x),a=3


```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(y**2*(3-x),x**2*(3+x)),(x,-5,5),(y,-5,5),line_color="blue",title='Strophoid:$y^2(a-x)=x^2(a+x)$')
```

3) Cissoid:y^2(a-x)=x^3,a=2

```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(y**2*(2-x),x**3),(x,-5,5),(y,-5,5),line_color="green",title='Cissoid:$y^2(a-x)=x^3$')
```

4) Lemniscate: a^2y^2=x^2(a^2-x^2),a=2

```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(2**2*y**2,x**2*(2**2-x**2)),(x,-5,5),(y,-5,5),line_color="red",title='Lemniscate: $a^2y^2=x^2(a^2-x^2)$')
```

5) Astroid: x^2(2/3)+y^2(2/3)=a^(2/3), a=2

```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(x**(2/3)+y**(2/3),2**(2/3)),(x,-5,5),(y,-5,5),line_color="purple",title='Astroid: $x^2(2/3)+y^2(2/3)=a^(2/3)$')
```

6) Folium of De-cartes : x^3+y^3+3axy, a=2

```python
from sympy import plot_implicit , symbols, Eq
x,y=symbols('x y')
p1=plot_implicit(Eq(x**3+y**3,3*2*x*y),(x,-5,5),(y,-5,5),line_color="green",title='$Folium of De-cartes: x^3+y^3+3axy')
```

7) Circle: x=acos(theta),y=bsin(theta), a=2, b=2

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
x=2*cos(theta)
y=2*sin(theta)
r=sqrt(x**2+y**2)
polar(theta,r,'g+')
show()
```
